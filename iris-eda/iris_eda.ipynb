**Importing python libraries**



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

**Loading the dataset**

```python
url = "https://bit.ly/4nejNue"
df = pd.read_csv(url)
```

```python
df.head(10)
```

```python
#Checking for the null values
df.isnull().sum()
```

**Basic Statistics (EDA)**

```python
df.describe()
```

```python
print("mean\n",df.mean(numeric_only=True))
print("median\n",df.median(numeric_only=True))
print("mode\n",df.mode(numeric_only=True))
```

**Visualize Histograms**

```python
df.hist(figsize=(10, 8), bins=15,color='blue')
plt.suptitle("Histograms of All Features")
plt.tight_layout()
plt.grid()
plt.show()
```
